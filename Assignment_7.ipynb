```python
#Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.
```

```python
'''The LookupError class in Python is a base class for exceptions that occur 
when attempting to access elements in a collection using invalid indices or keys.
It serves as a parent class for more specific lookup-related exception
classes like KeyError and IndexError.'''
```

```python
#1).KeyError :-

my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']  # 'd' key doesn't exist
except KeyError as e:
    print(f"KeyError: {e}")
```

```python
#2).IndexError :-

my_list = [10, 20, 30]

try:
    value = my_list[5]  # index 5 is out of range
except IndexError as e:
    print(f"IndexError: {e}")
```

```python
#Q5. Explain ImportError. What is ModuleNotFoundError?
```

```python
#ImportError:
'''ImportError is a generic exception that is raised when an imported module or package cannot be found or loaded. '''
```

```python
try:
    import non_existent_module
except ImportError:
    print("The module could not be imported.")
```

```python
#ModuleNotFoundError:
'''ModuleNotFoundError is a specific subclass of ImportError. 
It was introduced in Python 3.6 to provide a more informative error message when a module or package cannot be found.'''
```

```python
try:
    import non_existent_module
except ModuleNotFoundError:
    print("The specified module could not be found.")
```
